```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from datetime import datetime

# 设置中文字体
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus'] = False

def load_and_merge_data():
    """
    读取Excel文件并合并数据
    """
    try:
        # 读取两个工作表
        data1 = pd.read_excel('商品销售数据.xlsx', sheet_name='信息表')
        data2 = pd.read_excel('商品销售数据.xlsx', sheet_name='销售数据表')

        print("商品信息表前5行数据:")
        print(data1.head())
        print("\n销售数据表前5行数据:")
        print(data2.head())

        # 合并数据（右连接）
        data_all = pd.merge(data1, data2, on='商品编号', how='right')

        print("\n合并后的数据前5行:")
        print(data_all.head())
        print(f"\n数据形状: {data_all.shape}")

        return data_all
    except FileNotFoundError:
        print("错误：请确保'商品销售数据.xlsx'文件在当前目录下")
        return None
    except Exception as e:
        print(f"读取文件时发生错误: {e}")
        return None

def calculate_monthly_sales(data):
    """
    统计每月的销售金额
    """
    try:
        # ① 销售金额 = 商品销售价 * 订单数量
        # 检查商品销售价列的数据类型
        if not pd.api.types.is_string_dtype(data['商品销售价']):
            # 如果已经是数值类型，直接使用
            data['商品销售价'] = data['商品销售价'].astype(float)
        else:
            # 如果是字符串类型，去除¥符号并转换为数值
            data['商品销售价'] = data['商品销售价'].str.replace('¥', '').astype(float)

        data['销售金额'] = data['商品销售价'] * data['订单数量']

        # ② 从订单日期提取月份信息
        data['订单日期'] = pd.to_datetime(data['订单日期'])
        data['月份'] = data['订单日期'].dt.month

        # 按月份统计销售金额
        monthly_sales = data.groupby('月份')['销售金额'].sum().reset_index()

        print("\n每月销售金额统计:")
        print(monthly_sales)

        return data, monthly_sales
    except Exception as e:
        print(f"计算月度销售时发生错误: {e}")
        return None, None

def plot_monthly_sales_trend(monthly_sales):
    """
    绘制每月销售金额折线图
    """
    if monthly_sales is None:
        print("错误：无有效的月度销售数据")
        return

    plt.figure(figsize=(12, 6))

    # 提取月份和销售金额数据
    months = monthly_sales['月份']
    sales_amount = monthly_sales['销售金额']

    # 绘制折线图
    plt.plot(months, sales_amount, marker='o', linewidth=2, markersize=8,
             color='#FF6B6B', markerfacecolor='#4ECDC4')

    # 设置图表标题和坐标轴标签
    plt.title('商品每月销售金额变化趋势', fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('月份', fontsize=12)
    plt.ylabel('销售金额（元）', fontsize=12)

    # 设置x轴和y轴刻度
    plt.xticks(range(1, 13), [f'{i}月' for i in range(1, 13)])

    # 格式化y轴标签（添加千分位分隔符）
    plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))

    # 添加网格
    plt.grid(True, alpha=0.3)

    # 在每个数据点上添加数值标签
    for i, (month, sales) in enumerate(zip(months, sales_amount)):
        plt.annotate(f'{sales:,.0f}', (month, sales),
                    textcoords="offset points", xytext=(0,10),
                    ha='center', fontsize=9)

    plt.tight_layout()
    plt.show()

def analyze_category_sales(data):
    """
    分析各商品大类的月度销售额
    """
    # 按照月份和商品大类分组汇总
    category_monthly_sales = data.groupby(['月份', '商品大类'])['销售金额'].sum().reset_index()

    print("\n各商品大类月度销售金额:")
    print(category_monthly_sales)

    return category_monthly_sales

def plot_category_sales_comparison(category_sales):
    """
    绘制不同商品大类的月度销售额柱状图
    """
    if category_sales is None or category_sales.empty:
        print("错误：无有效的品类销售数据")
        return

    # 获取唯一的商品大类
    categories = category_sales['商品大类'].unique()
    months = range(1, 13)

    plt.figure(figsize=(14, 8))

    # 设置柱状图宽度和位置
    bar_width = 0.25
    x_pos = np.arange(len(months))

    # 为每个商品大类绘制柱状图
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FECA57']

    for i, category in enumerate(categories):
        # 获取该品类每个月的销售数据
        category_data = []
        for month in months:
            monthly_data = category_sales[
                (category_sales['月份'] == month) &
                (category_sales['商品大类'] == category)
            ]
            if not monthly_data.empty:
                category_data.append(monthly_data['销售金额'].iloc[0])
            else:
                category_data.append(0)

        # 绘制柱状图
        plt.bar(x_pos + i * bar_width, category_data, bar_width,
                label=category, color=colors[i % len(colors)], alpha=0.8)

    # 设置图表属性
    plt.title('商品大类月度销售金额对比图',
              fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('月份', fontsize=12)
    plt.ylabel('销售金额（元）', fontsize=12)

    # 设置x轴刻度
    plt.xticks(x_pos + bar_width, [f'{i}月' for i in months])

    # 格式化y轴标签
    plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))

    # 添加图例
    plt.legend(title='商品大类', bbox_to_anchor=(1.05, 1), loc='upper left')

    # 添加网格
    plt.grid(True, alpha=0.3, axis='y')

    plt.tight_layout()
    plt.show()

def analyze_subcategory_sales(data):
    """
    分析各商品小类的月度销售额
    """
    # 按照月份和商品小类分组汇总
    subcategory_monthly_sales = data.groupby(['月份', '商品大类', '商品小类'])['销售金额'].sum().reset_index()

    print("\n各商品小类月度销售金额:")
    print(subcategory_monthly_sales.head(10))  # 只显示前10行，避免输出过长

    # 统计每个小类的总销售额
    subcategory_totals = data.groupby(['商品大类', '商品小类'])['销售金额'].sum().sort_values(ascending=False)
    print("\n商品小类销售总额排名:")
    print(subcategory_totals)

    return subcategory_monthly_sales

def plot_subcategory_sales_trend(subcategory_sales, top_n=8):
    """
    绘制Top N商品小类的月度销售趋势图
    """
    if subcategory_sales is None or subcategory_sales.empty:
        print("错误：无有效的小类销售数据")
        return

    # 计算每个小类的总销售额并排序
    subcategory_totals = subcategory_sales.groupby(['商品大类', '商品小类'])['销售金额'].sum().sort_values(ascending=False)

    # 选择Top N的小类
    top_subcategories = subcategory_totals.head(top_n).index

    plt.figure(figsize=(14, 8))

    # 为每个Top小类绘制趋势线
    colors = plt.cm.Set3(np.linspace(0, 1, len(top_subcategories)))

    for i, (category, subcategory) in enumerate(top_subcategories):
        # 筛选该小类的数据
        sub_data = subcategory_sales[
            (subcategory_sales['商品大类'] == category) &
            (subcategory_sales['商品小类'] == subcategory)
        ]

        if not sub_data.empty:
            # 确保包含所有月份的数据
            all_months = pd.DataFrame({'月份': range(1, 13)})
            sub_data_complete = pd.merge(all_months, sub_data, on='月份', how='left')
            sub_data_complete['销售金额'] = sub_data_complete['销售金额'].fillna(0)

            # 绘制折线图
            plt.plot(sub_data_complete['月份'], sub_data_complete['销售金额'],
                    marker='o', linewidth=2, markersize=6,
                    label=f'{category}-{subcategory}', color=colors[i])

    plt.title(f'Top {top_n} 商品小类月度销售趋势', fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('月份', fontsize=12)
    plt.ylabel('销售金额（元）', fontsize=12)
    plt.xticks(range(1, 13), [f'{i}月' for i in range(1, 13)])
    plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))
    plt.grid(True, alpha=0.3)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

def plot_subcategory_heatmap(subcategory_sales, top_n=10):
    """
    绘制商品小类月度销售热力图
    """
    if subcategory_sales is None or subcategory_sales.empty:
        print("错误：无有效的小类销售数据")
        return

    # 计算每个小类的总销售额并排序
    subcategory_totals = subcategory_sales.groupby(['商品大类', '商品小类'])['销售金额'].sum().sort_values(ascending=False)
    top_subcategories = subcategory_totals.head(top_n).index

    # 创建热力图数据
    heatmap_data = []
    subcategory_labels = []

    for category, subcategory in top_subcategories:
        sub_data = subcategory_sales[
            (subcategory_sales['商品大类'] == category) &
            (subcategory_sales['商品小类'] == subcategory)
        ]

        # 确保包含所有月份的数据
        monthly_sales = []
        for month in range(1, 13):
            month_data = sub_data[sub_data['月份'] == month]
            if not month_data.empty:
                monthly_sales.append(month_data['销售金额'].iloc[0])
            else:
                monthly_sales.append(0)

        heatmap_data.append(monthly_sales)
        subcategory_labels.append(f'{category}\n{subcategory}')

    heatmap_data = np.array(heatmap_data)

    # 绘制热力图
    plt.figure(figsize=(12, 8))
    im = plt.imshow(heatmap_data, cmap='YlOrRd', aspect='auto')

    # 设置坐标轴
    plt.yticks(range(len(subcategory_labels)), subcategory_labels)
    plt.xticks(range(12), [f'{i}月' for i in range(1, 13)])

    # 添加颜色条
    cbar = plt.colorbar(im)
    cbar.set_label('销售金额（元）')

    # 添加数值标签
    for i in range(len(subcategory_labels)):
        for j in range(12):
            text = plt.text(j, i, f'{heatmap_data[i, j]:.0f}',
                           ha="center", va="center", color="black", fontsize=8)

    plt.title('商品小类月度销售热力图', fontsize=16, fontweight='bold', pad=20)
    plt.tight_layout()
    plt.show()

def generate_analysis_report(data, monthly_sales, category_sales, subcategory_sales):
    """
    生成数据分析报告
    """
    print("=" * 60)
    print("                   商品销售数据分析报告")
    print("=" * 60)

    # 基本统计信息
    total_sales = data['销售金额'].sum()
    total_orders = data['订单数量'].sum()
    avg_order_value = total_sales / total_orders

    print(f"\n1. 基本统计信息:")
    print(f"   - 总销售金额: {total_sales:,.2f} 元")
    print(f"   - 总订单数量: {total_orders:,.0f} 件")
    print(f"   - 平均客单价: {avg_order_value:,.2f} 元/件")

    # 月度分析
    best_month = monthly_sales.loc[monthly_sales['销售金额'].idxmax()]
    worst_month = monthly_sales.loc[monthly_sales['销售金额'].idxmin()]

    print(f"\n2. 月度销售分析:")
    print(f"   - 销售最佳月份: {int(best_month['月份'])}月, 销售额: {best_month['销售金额']:,.2f} 元")
    print(f"   - 销售最差月份: {int(worst_month['月份'])}月, 销售额: {worst_month['销售金额']:,.2f} 元")

    # 品类分析
    category_totals = data.groupby('商品大类')['销售金额'].sum().sort_values(ascending=False)

    print(f"\n3. 商品大类销售排名:")
    for i, (category, sales) in enumerate(category_totals.items(), 1):
        percentage = (sales / total_sales) * 100
        print(f"   {i}. {category}: {sales:,.2f} 元 ({percentage:.1f}%)")

    # 小类分析
    subcategory_totals = data.groupby(['商品大类', '商品小类'])['销售金额'].sum().sort_values(ascending=False)

    print(f"\n4. Top 5 商品小类销售排名:")
    for i, ((category, subcategory), sales) in enumerate(subcategory_totals.head(5).items(), 1):
        percentage = (sales / total_sales) * 100
        print(f"   {i}. {category}-{subcategory}: {sales:,.2f} 元 ({percentage:.1f}%)")

def main():
    """
    主函数：执行完整的数据分析流程
    """
    print("开始商品销售数据分析...")

    # 1. 获取数据
    data_all = load_and_merge_data()
    if data_all is None:
        return

    # 2. 分类统计数据
    data_with_sales, monthly_sales = calculate_monthly_sales(data_all)
    if data_with_sales is None:
        return

    # 3. 绘制每月销售趋势图
    plot_monthly_sales_trend(monthly_sales)

    # 4. 分析商品大类销售情况
    category_sales = analyze_category_sales(data_with_sales)
    plot_category_sales_comparison(category_sales)

    # 5. 分析商品小类销售情况（新增）
    subcategory_sales = analyze_subcategory_sales(data_with_sales)
    plot_subcategory_sales_trend(subcategory_sales, top_n=8)
    plot_subcategory_heatmap(subcategory_sales, top_n=10)

    # 6. 生成分析报告
    generate_analysis_report(data_with_sales, monthly_sales, category_sales, subcategory_sales)

    print("\n数据分析完成！")

if __name__ == "__main__":
    main()
```
